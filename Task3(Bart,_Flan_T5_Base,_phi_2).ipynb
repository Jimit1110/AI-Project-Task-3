# **Bart base**

```python
import gradio as gr
from transformers import BartTokenizer, BartForConditionalGeneration
```

```python
tokenizer = BartTokenizer.from_pretrained("facebook/bart-base")
model = BartForConditionalGeneration.from_pretrained("facebook/bart-base")
```

```python
def generate_book_name(query):
    input_text = "generate research book name: " + query + " </s>"
    input_ids = tokenizer.encode(input_text, return_tensors="pt")

    outputs = model.generate(input_ids=input_ids, max_length=100, num_beams=5, early_stopping=True)
    generated_name = tokenizer.decode(outputs[0], skip_special_tokens=True)

    return generated_name

def generate_enhanced_fact(fact):
    input_text = "Enhance the following fact: " + fact + " </s>"
    input_ids = tokenizer.encode(input_text, return_tensors="pt")

    outputs = model.generate(input_ids=input_ids, max_length=100, num_beams=5, early_stopping=True)
    enhanced_fact = tokenizer.decode(outputs[0], skip_special_tokens=True)

    return enhanced_fact
```

```python
output_text = gr.Textbox()
gr.Interface(generate_book_name, "textbox", output_text, title="BART for Research Book Name Generation",
             description="Enter a query to generate a research book name.").launch()

output_fact = gr.Textbox()
gr.Interface(generate_enhanced_fact, "textbox", output_fact, title="BART for Enhanced Fact Generation",
             description="Enter a fact to generate an enhanced fact.").launch()
```
